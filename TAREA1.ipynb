```
print("programa de peso de juguetes por pedido")

def peso (payaso, mesa) :
  res1 = payaso * 150
  res2 = mesa * 170
  res3 = res1 + res2 
  print("el peso de los payasos es de:", res1,"gr")
  print("el peso de los juegos de mesa es de:", res2,"gr")
  print("el peso total del pedido es de:", res3,"gr")

  if res3 > 10000 :
    print("error-exeso de peso")
    res4 = res3 - 10000
    print("exeso de peso = ", res4,"gr")
  else :
    print("pedido exitoso")

a = int (input("ingrese la cantidad de payasos:"))
b = int (input("ingrese la cantidad de juegos de mesa:"))

peso (a, b)

print("fin del programa")
```

EJERCICIO 9


```
print("programa factorial") 

def factorial(entero):
  i = 1
  fact = 1 
  while i < entero  :
    i = i + 1
    fact = fact * i 
    res = fact

  print("resultado del factorial =", res)


a = int (input("ingrese un numero entero:"))

factorial(a)
print("fin del programa")
```
